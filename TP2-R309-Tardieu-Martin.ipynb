```python
import tkinter as tk
from tkinter import simpledialog
from PIL import Image, ImageTk

class ImageObject:
    def __init__(self, path, size, color):
        self.path = path
        self.size = size
        self.color = color
        self.name = path 
        self.label_id = None

def on_image_click(event, info):
    global selected_object
    selected_object = {"id": None, "image": None, "info": info}
    x, y = event.x, event.y
    image = Image.open(info.path)
    resize_image = image.resize(info.size)
    img = ImageTk.PhotoImage(resize_image)

    selected_object["id"] = canva.create_image(x, y, image=img)
    selected_object["image"] = img
    canva.tag_bind(selected_object["id"], '<ButtonPress-1>', lambda event, info=selected_object: on_image_click2(event, info))
    canva.tag_bind(selected_object["id"], '<Button-3>', lambda event, info=selected_object: show_context_menu(event, info))
    info.label_id = canva.create_text(x, y + info.size[1] + 5, text=info.name, fill="white", anchor="n")
    image_objects.append({"id": selected_object["id"], "image": selected_object["image"], "info": selected_object["info"]})
    print(selected_object)

def on_image_click2(event, info):
    global selected_object
    selected_object = info
    print(info["id"])

def on_mouse_drag(event):
    global selected_object
    if selected_object["id"]:
        x, y = event.x, event.y
        canva.coords(selected_object["id"], x, y)
        canva.coords(selected_object["info"].label_id, x, y + selected_object["info"].size[1] + 5)

def supprimer_image(info):
    global selected_object
    if info["id"]:
        canva.delete(info["id"])
        canva.delete(info["info"].label_id)
        image_objects.remove(info)
        selected_object = {"id": None, "image": None, "info": None}

def show_context_menu(event, info):
    context_menu = tk.Menu(root, tearoff=0)
    context_menu.add_command(label="Supprimer", command=lambda: supprimer_image(info))
    context_menu.add_command(label="Changer le nom", command=lambda: change_name(info))
    context_menu.post(event.x_root, event.y_root)

def change_name(info):
    new_name = simpledialog.askstring("Changer le nom", "Nouveau nom:", initialvalue=info.name)
    if new_name:
        info.name = new_name
        canva.itemconfig(selected_object["info"].label_id, text=new_name)

root = tk.Tk()
canva = tk.Canvas(bg="#222", height=800, width=1200)
canva.pack()
canva.create_rectangle(0, 0, 150, 800, fill="white")

image_objects = []
selected_object = {"id": None, "image": None, "info": None}

images_info = [
    ImageObject("pc.png", (100, 100), "#586F6B"),
    ImageObject("telephone.png", (50, 100), "#7F9183"),
    ImageObject("routeur.png", (100, 75), "#B8B8AA"),
    ImageObject("switch.png", (100, 50), "#CFC0BD"),
]

for i, image_info in enumerate(images_info):
    x_position = 75
    y_position = i * 200 + 100
    fill_color = image_info.color

    canva.create_rectangle(0, i * 200, 150, i * 200 + 200, fill=fill_color)

    image = Image.open(image_info.path)
    resize_image = image.resize(image_info.size)
    img = ImageTk.PhotoImage(resize_image)

    image_id = canva.create_image(x_position, y_position, image=img)
    canva.tag_bind(image_id, '<ButtonPress-1>', lambda event, info=image_info: on_image_click(event, info))
    canva.tag_bind(image_id, '<Button-3>', lambda event, info=image_info: show_context_menu(event, info))

    image_info.label_id = canva.create_text(x_position, y_position + image_info.size[1] + 5, text=image_info.name, fill="white", anchor="n")

    image_objects.append({"id": image_id, "image": img, "info": image_info})

canva.bind('<B1-Motion>', on_mouse_drag)

root.mainloop()
```
